# Python Charts Made Easy: Unlocking Data Insights with Visualization

```python
###Line Plot

#import libraries
import matplotlib.pyplot as plt
import pandas as pd

# Example data (stock prices)
data = {
 'Date': ['2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01'],
 'Stock Price': [150, 160, 155, 170]
}

df = pd.DataFrame(data)

# Create a line chart
plt.plot(df['Date'], df['Stock Price'], marker='o')

# Customize the plot
plt.title('Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Stock Price')

# Show the plot
plt.show()
```

```python
### Bar plot

#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example data (car sales)
data = {
 'Car Model': ['Sedan', 'SUV', 'Truck'],
 'Sales': [250, 300, 180]
}

df = pd.DataFrame(data)

# Create a bar chart
sns.barplot(x='Car Model', y='Sales', data=df)

# Customize the plot
plt.title('Car Sales by Model')
plt.xlabel('Car Model')
plt.ylabel('Sales')

# Show the plot
plt.show()
```

```python
### Scatter Plot

#Import libraries
import matplotlib.pyplot as plt
import pandas as pd

# Example data (student data)
data = {
 'Hours Studied': [2, 3, 1.5, 4, 5, 2.5, 3.5, 3.2, 4.2, 5, 7, 6],
 'Exam Score': [75, 85, 60, 90, 95, 80, 88, 90, 68, 75, 37, 89]
}

df = pd.DataFrame(data)

# Create a scatter plot
plt.scatter(df['Hours Studied'], df['Exam Score'])

# Customize the plot
plt.title('Scatter Plot of Hours Studied vs. Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')

# Show the plot
plt.show()
```

```python
### Histogram

#Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example data (student ages)
data = {
 'Age': [19, 22, 20, 23, 21, 25, 24, 20, 22, 18, 19, 21]
}

df = pd.DataFrame(data)

# Create a histogram
sns.histplot(df['Age'], bins=5, kde=True)

# Customize the plot
plt.title('Histogram of Student Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
### Pie chart

#Import libraries
import matplotlib.pyplot as plt

# Example data (expense categories and percentages)
categories = ['Salaries', 'Marketing', 'Utilities', 'Supplies']
expenses = [40, 25, 15, 20]  # Percentages

# Create a pie chart
plt.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=140)

# Customize the plot
plt.title('Expense Distribution')

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
### Box plot

#Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example data (housing prices by neighborhood)
data = {
 'Neighborhood': ['A', 'B', 'C', 'D', 'A', 'B', 'C', 'D'],
 'Price': [350000, 400000, 320000, 450000, 300000, 420000, 330000, 460000]
}

df = pd.DataFrame(data)

# Create a box plot
sns.boxplot(x='Neighborhood', y='Price', data=df)

# Customize the plot
plt.title('Housing Prices by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Price')

# Show the plot
plt.show()
```

```python
### heatamp

#Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example data (correlation matrix)
data = {
 'Asset 1': [1.0, 0.5, -0.2, 0.8],
 'Asset 2': [0.5, 1.0, 0.3, -0.1],
 'Asset 3': [-0.2, 0.3, 1.0, 0.6],
 'Asset 4': [0.8, -0.1, 0.6, 1.0]
}

df = pd.DataFrame(data)

# Create a heatmap
sns.heatmap(df, annot=True, cmap='coolwarm')

# Customize the plot
plt.title('Correlation Matrix of Portfolio Assets')

# Show the plot
plt.show()
```
